# 3. Bulk movies

Select the runs and call one simple function.

```python
from pathlib import Path

from bulk_movies import make_movies

PARAM_BASE = "params_four_phases"
DATA_FOLDER = Path("data") / PARAM_BASE
MOVIE_FOLDER = Path("movies") / PARAM_BASE

RUN_IDS = [0, 1, 2, 3]
MODE = "preview"       # "preview" or "hq"
MAX_PARALLEL = 4
SKIP_EXISTING = True
```

```python
movie_results = make_movies(
    data_folder=DATA_FOLDER,
    param_base=PARAM_BASE,
    run_ids=RUN_IDS,
    movie_folder=MOVIE_FOLDER,
    mode=MODE,
    max_parallel=MAX_PARALLEL,
    skip_existing=SKIP_EXISTING,
)
```
