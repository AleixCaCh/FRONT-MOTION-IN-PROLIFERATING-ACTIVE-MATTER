# 1. Compile and run the bulk simulations

```python
from pathlib import Path

from bulk_runner import compile_code, read_run_ids, run_parallel

MAX_PARALLEL = 4
SKIP_EXISTING = False
```

## Main model: inherited newborn orientation

```python
C_FILE = Path("abm_bulk_validation.c")
EXE_FILE = Path("abm_bulk_validation.exe")
PARAM_FILE = Path("params_four_phases.txt")
DATA_FOLDER = Path("data") / PARAM_FILE.stem

RUN_IDS = read_run_ids(PARAM_FILE)
print(RUN_IDS)
```

```python
compile_code(C_FILE, EXE_FILE)

main_results = run_parallel(
    exe_file=EXE_FILE,
    param_file=PARAM_FILE,
    run_ids=RUN_IDS,
    data_folder=DATA_FOLDER,
    max_parallel=MAX_PARALLEL,
    skip_existing=SKIP_EXISTING,
)
```

## Alternative model: random newborn orientation

```python
C_FILE_ALT = Path("abm_bulk_validation_non_inheritance.c")
EXE_FILE_ALT = Path("abm_bulk_validation_non_inheritance.exe")
PARAM_FILE_ALT = Path("params_four_phases_altered.txt")
DATA_FOLDER_ALT = Path("data") / PARAM_FILE_ALT.stem

RUN_IDS_ALT = read_run_ids(PARAM_FILE_ALT)
print(RUN_IDS_ALT)
```

```python
compile_code(C_FILE_ALT, EXE_FILE_ALT)

alternative_results = run_parallel(
    exe_file=EXE_FILE_ALT,
    param_file=PARAM_FILE_ALT,
    run_ids=RUN_IDS_ALT,
    data_folder=DATA_FOLDER_ALT,
    max_parallel=MAX_PARALLEL,
    skip_existing=SKIP_EXISTING,
)
```
